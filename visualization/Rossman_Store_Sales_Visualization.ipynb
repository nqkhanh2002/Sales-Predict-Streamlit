## Plotting functions 

```python
#catplot
def cat(df, x, y, k = 'strip'):
    sns.catplot(data=df,x = x ,y = y,  kind = k)

#catplot of larger size
def big_cat(x, y, df, k):
    fig, ax = plt.subplots(1, 1, figsize=(8, 6))
    sns.catplot(ax = ax, x = x, y = y, data=df, kind = k)
    plt.close(2)
    plt.show()
    
#date specific plotting
def dateplot(x, y, df, k):
    fig, ax = plt.subplots(1, 1, figsize=(80, 18))
    sns.catplot(ax = ax, x = x, y = y, data=df, kind = k)
    plt.close(2)
    plt.show()
```
